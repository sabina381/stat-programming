# 1>

```python
def my_function(mat):
    return mat - np.mean(mat, axis=1, keepdims=True)
```

- `np.mean` 함수를 사용해 row별 mean을 구하고, 원래 행렬에서 row mean 행렬을 빼는 방식으로 구현하였다.
    - `axis=1` 로 지정해 row별 mean을 구하였다.
    - `keepdims=True` 옵션을 통해 row의 차원을 유지하도록 했다. 원래 행렬에서 각 row의 mean을 빼야하기 때문에 차원을 유지하도록 했다.(옵션을 지정하지 않으면 각 row의 mean을 1차원 vector 형태로 출력한다.)

- numpy에서는 행렬끼리 뺄 때, 차원이 맞지 않아도 broadcasting하여 계산해주는 기능이 있다. 여기서 원래 행렬 mat(nxm)에서 row mean 행렬(nx1)을 뺄 때, row mean 행렬을 원래 행렬에 맞춰 broadcasting해서 뺀다.

```python
A = np.array([[1,2,5,1], [3,4,1,0]])
my_function(A)
```

```python
A = np.array([[1,2,5,1], [3,4,1,0]])
my_function(A)
```

# 5>

```python
def my_function(row, col):
    A = np.zeros([row-2, col-2])
    A = np.pad(A, pad_width=1, mode='constant', constant_values=1)

    return A
```

- input: 만들 행렬의 row, col 개수
- `np.zeros`함수로 우선 가장자리 상하좌우 한 줄씩 뺀 (row-2, col-2)차원의 0 행렬을 만든다.
- `np.pad`함수로 가장자리 상하좌우에 '1'값을 추가한다.
    - `pad_width=1`: 가장자리에 1줄을 추가한다. (size 1의 패딩을 한다)
    - `mode='constant'`: 가장자리 상수 타입의 값을 패딩한다.
    - `constant_values=1`: 상수값 '1'을 패딩한다.

```python
my_function(10,10)
```

# 6>

```python
def my_function(lst, window_size):
    result = []
    for i in range(len(lst)-window_size+1):
        result.append(np.round(np.mean(lst[i:i+window_size]),3))

    return(result)
```

- input: lst(계산할 리스트), window_size(몇 개씩의 평균을 계산할 것인지)
- 우선 결과를 담을 리스트를 만들고, for문을 이용해 `(len(lst)-window\_size+1)`만큼 반복하면서 한 칸씩 이동하며 window_size만큼의 평균을 구하고 `.append`메서드를 이용해 결과 리스트에 값을 추가한다. `np.round`함수를 이용해 소수점은 3자리까지만 표시했다.

```python
A = [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
my_function(A, 3)
```

## (+) sample과 똑같이 출력하도록 한 함수

```python
def my_function(lst, window_size):
    i_str = ' '.join([str(i) for i in range(len(lst)-window_size+1)]) # 반복 횟수 출력
    print(f"i {i_str}")
    print("= "*10)
    print("Input", lst) # input 리스트 출력

    for i in range(len(lst)-window_size+1):
        lst_str = ''
        temp = [str(lst[j]) for j in range(i, i+window_size)] # list complication으로 계산에 쓰인 원래 값들의 리스트 만들기
        lst_str = '+'.join(temp) # join 메서드로 각 리스트의 원소들을 '+'로 연결한 하나의 문자열 만들기
        print(f"y{i+1} {np.round(np.mean(lst[i:i+window_size]),3)} = ({lst_str})/{window_size}") # 계산한 평균과 식을 함께 출력
```

- 위와 계산하는 것은 같고, 결과와 식까지 sample과 같이 예쁘게 출력하는 함수이다..

```python
A = [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
my_function(A, 3)
```

```python
Y = np.array([[[  0, 157,   1],
            [  1, 167,   1],
            [  0, 130,   1],
            [  1, 145,   1],
            [  2, 170,   1]],

        [[  1, 250,   0],
            [  3, 260,   0],
            [  2, 263,   0],
            [  1, 273,   0],
            [  0, 290,   0]],

        [[  0, 301,   0],
            [  0, 120,   0],
            [  0,  30,   0],
            [  1, 100,   0],
            [  3, 502,   0]]])
```

- 문제에 주어진대로 첫 번째 차원에 id, 두 번째 차원에 year, 세 번째 차원에 순서대로 N, salary, sports car 정보를 표시하도록 3차원 array를 만들었다.

## c>

```python
Y = np.stack([N, Salary, Sports],axis=2)
```

- 원래 N, Salary, Sports 행렬의 row(첫 번째 차원)는 id, column(두 번째 차원)은 year을 표현하고 있었기 때문에, `np.stack` 함수를 사용해 이들을 세 번째 차원으로 join했다.
- `np.stack`함수는 입력받은 행렬을 새로운 축으로 join한다. 몇 번째 축으로 join할지는 `axis` 옵션으로 지정한다.
